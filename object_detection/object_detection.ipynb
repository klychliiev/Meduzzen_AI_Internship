Define varibales to access the dataset items

```python
NP_DATASET_PATH = "nova_poshta_dataset/"
POSTS_PHOTOS = 'Фото поштоматів'
PARCELS_TYPES = "Parcels_types.jpg"
POSTALBOX_TEMPLATES = "nova_poshta_postalbox_templates.xlsx"
CELLS_SIZES = "Розміри комірок.jpg"
```

Parcel machines pictures

```python
import os 
```

```python
pictures = os.path.join(NP_DATASET_PATH, POSTS_PHOTOS)
pictures_list = os.listdir(pictures)
```

```python
# number of pictures of postal machines
len(pictures_list)
```

```python
from IPython.display import display, Image 

image_path = os.path.join(pictures, pictures_list[0])
img = Image(image_path)

# display the first imahe of a parcel machine from the folder "Фото поштоматів"
display(img)
```

```python
# sizes of the first 10 images 

import cv2 

for i in range(10):
    im = cv2.imread(os.path.join(pictures, pictures_list[i]))
    print(im.shape)
```

Cells sizes

```python
cells_sizes = os.path.join(NP_DATASET_PATH, CELLS_SIZES)
cells_image = Image(cells_sizes)
display(cells_image)
```

Parcels types

```python
parcels_types = os.path.join(NP_DATASET_PATH, PARCELS_TYPES)
parcels_image = Image(parcels_types)
display(parcels_image)
```
